# **Project and Dataset Summary:**
In this project, we analyzed a bank's customer dataset to understand the factors contributing to customer churn (i.e., customers exiting).
The Dataset includes Account information for 10,000 customers at a European bank, including details on their credit score, balance, products, and whether they have churned.

# **Problem Statement:**
We need to explore and analyze the bank's customer data to discover the key factors responsible for customer churn.

```python
desc = df.describe(include= 'all')
```

```python
Numerical_Columns = desc.drop(columns = ['Surname', 'Geography', 'Gender','HasCrCard','IsActiveMember','Exited']).columns.values
```

```python
Numerical_Columns
```

```python
Categorical_Columns = ['Surname', 'Geography', 'Gender','HasCrCard','IsActiveMember','Exited']
```

```python
Categorical_Columns
```

# **Univariate Analysis**

```python
plt.figure(figsize=(18,9))
plt.suptitle('Univariate analysis: Distribution plots', fontsize=20)
for i in range(0, len(Numerical_Columns)):
  plt.subplot(3,3,i+1)
  sns.kdeplot(x=df[Numerical_Columns[i]],fill=True)
```

```python
df['Exited'].value_counts()
```

**2037** Customers have Churned or Exited.


```python
plt.figure(figsize=(20,9))
plt.suptitle('Analysis of variables during Exited and Not Exited Customers', fontsize=20)
for i in range(0, len(Numerical_Columns)):
  plt.subplot(3,3,i+1)
  sns.kdeplot(x=df[Numerical_Columns[i]], hue =df['Exited'] , fill=False)
```

# **Conclusions:**
1. Churn rate is high when the **Credit** **Score** is between **600** to **700**.
2. Churn rate is high when the **Age** of the Customers is near to **50**.
3. Churn rate is high when the **Tenure** hits **1** **Year** and when the Tenure is **8**-**9** Years.
4. Churn rate is high when the **Account** **Balance** is **0** and when the **Account** **Balance** is between **100000** and **150000**.
5. Customers having less **Number** **of** **Products** have Churned **more**.
6. The **more** the Estimated Salary is the **more** the Churn rate.

```python
plt.figure(figsize=(14,6))
for i in range(0, len(Categorical_Columns)):
  ax=sns.countplot(x=df[Categorical_Columns[i]])
  plt.xlabel(Categorical_Columns[i])
  plt.xticks(rotation=90)
  plt.show()
```

# **Conclusions:**

1. **Most** of the Customers are from **France**.
2. **More** than **5000** Customers are **Male**.
3. **7000** Customers have **Credit** **Card**.
4. **More** than **5000** Customers are **Active** **Members**.

# **Multivariate Analysis**

```python
corr = df.drop(columns=['CustomerId', 'Surname', 'Geography', 'Gender']).corr()
```

```python
plt.figure(figsize = (16,10))
sns.heatmap(corr, annot = True)
```

From the above heatmap we can conclude that the most corrilated variables with our target variable i.e. Churn are Age, Balance and Estimated Salary.

# **Final** **Conclusions**:

1. Age, Balance, and Estimated Salary show a stronger correlation with churn, as seen in the correlation heatmap.

2. Middle-aged customers (around 50 years old) and customers with a credit score between 600–700 are more likely to churn.

3. Customers with 0 balance or those with balances ranging between 100,000 to 150,000 show a higher churn tendency.

4. Customers having fewer products (typically 1) are more prone to leave, indicating lower engagement or fewer services availed.

5. While tenure generally doesn’t show strong correlation, a spike in churn is observed in customers with 1 year and 8–9 years of tenure.

6. Most customers are from France, and churn is highest in France and Germany, indicating geographic patterns.

7. Female customers churn more than males, despite the overall higher count of male customers.

8. Customers who have a credit card or are active members are less likely to churn, suggesting these are retention-driving factors.
